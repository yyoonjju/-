```python
regions = ['강원도',
'경기도',
'경상남도',
'경상북도',
'광주광역시',
'대구광역시',
'대전광역시',
'부산광역시',
'서울특별시',
'세종특별자치시',
'울산광역시',
'인천광역시',
'전라남도',
'전라북도',
'제주특별자치도',
'충청남도',
'충청북도']
df = pd.DataFrame(columns=pd.DataFrame(evcs.loc[evcs.주소.str.contains(regions[0]),:].sum()).T.columns)
for region in regions:
    df = pd.concat([df, pd.DataFrame(evcs.loc[evcs.주소.str.contains(region),:].sum()).T])
df=df.reset_index()
```
